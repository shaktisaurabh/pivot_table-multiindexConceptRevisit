```python
import pandas as pd

# Sample DataFrame with MultiIndex
index = pd.MultiIndex.from_tuples([('A', 1), ('A', 2), ('B', 1), ('B', 2)], names=['Letter', 'Number'])
columns = pd.MultiIndex.from_tuples([('X', 'value'), ('Y', 'value')], names=['Axis', 'Metric'])
df = pd.DataFrame([[10, 20], [30, 40], [50, 60], [70, 80]], index=index, columns=columns)

print("DataFrame with MultiIndex:")
print(df)

# Swap levels of the MultiIndex along columns
df_swapped = df.reorder_levels([1, 0], axis=1)

print("\nDataFrame with swapped levels of MultiIndex along columns:")
print(df_swapped)
```

```python
import pandas as pd

# Sample DataFrame with MultiIndex
index = pd.MultiIndex.from_tuples([('A', 1), ('A', 2), ('B', 1), ('B', 2)], names=['Letter', 'Number'])
columns = pd.Index(['Value'])
data = [[10], [20], [30], [40]]
df = pd.DataFrame(data, index=index, columns=columns)

print("Original DataFrame:")
print(df)

# Select rows where 'Letter' is 'A' without dropping the level
result = df.xs('A', level='Letter', drop_level=False)

print("\nResult after cross-section selection with drop_level=False:")
print(result)
```

```python
import pandas as pd

# Sample DataFrame with MultiIndex
index = pd.MultiIndex.from_tuples([('A', 1), ('A', 2), ('B', 1), ('B', 2)], names=['Letter', 'Number'])
columns = pd.Index(['Value'])
data = [[10], [20], [30], [40]]
df = pd.DataFrame(data, index=index, columns=columns)

print("Original DataFrame:")
print(df)

# Select rows where 'Letter' is 'A' without dropping the level
result = df.xs('A', level='Letter', drop_level=True)

print("\nResult after cross-section selection with drop_level=False:")
print(result)
#as you can see drop_level=True drops the selected level from the result.
```
